```python
from matplotlib import pyplot as plt
import numpy as np
from numpy import arange
from pylab import plot,xlabel,ylabel,show


NAt = 25000 

NRn = 0
NPb211 = 0
NBi215 = 0
NBi211 = 0
NTl = 0
NPb207 = 0
Nalpha_decay = 0
NR_decay = 0
Nbeta_decay = 0
NZ_decay = 0
alpha_energy = 0
R_energy = 0
beta_energy = 0
Z_energy = 0
tau_At = 560
tau_Bi215 = 7.6 * 60
tau_Rn = 40
tau_Pb211 = 36.1 * 60
tau_Bi211 = 2.14 * 60
tau_Tl = 4.77 * 60
h = 1.0
p_At = 1-2**(-h/tau_At)
p_Rn = 1-2**(-h/tau_Rn)
p_Pb211 = 1-2**(-h/tau_Pb211)
p_Bi215 = 1-2**(-h/tau_Bi215)
p_Bi211 = 1-2**(-h/tau_Bi211)
p_Tl = 1 - 2**(-h/tau_Tl)
tmax = 20000                


tpoints = arange(0.0,tmax,h) 
Atpoints = []
Rnpoints = []
Pb211points = []
Bi215points = []
Bi211points = []
Tlpoints = [] 
Pb207points = [] 

for t in tpoints:
    Atpoints.append(NAt)
    Rnpoints.append(NRn)
    Pb211points.append(NPb211)
    Bi215points.append(NBi215)
    Bi211points.append(NBi211)
    Tlpoints.append(NTl) 
    Pb207points.append(NPb207)

    decay_At = 0
    decay_Rn = 0
    decay_Pb211 = 0
    decay_Bi215 = 0
    decay_Bi211 = 0
    decay_Tl = 0
    decay_Pb207 = 0
    for i in range(NAt): 
        if np.random.random() < p_At:
            decay_At += 1
            if np.random.random() > .3:
                NBi215 += decay_At
                NAt -= decay_At
                Nalpha_decay +=  decay_At
            else:
                NRn += decay_At
                NAt -= decay_At
                Nbeta_decay += decay_At
    for a in range(NRn):
        if np.random.random() < p_Rn:
            decay_Rn+= 1
    NRn -= decay_Rn
    NPb211 += decay_Rn
    NZ_decay += decay_Rn
    for q in range(NBi215):
        if np.random.random() < p_Bi215:
            decay_Bi215 += 1
    NBi215 -= decay_Bi215
    NPb211 += decay_Bi215
    NR_decay+= decay_Bi215
    for w in range(NPb211):
        if np.random.random() < p_Pb211:
            decay_Pb211 += 1
    NPb211 -= decay_Pb211
    NBi211 += decay_Pb211
    Nbeta_decay += decay_Pb211
    for k in range(NBi211): 
        if np.random.random() < p_Bi211:
            decay_Bi211 += 1
            if np.random.random() > .03:
                NTl += decay_Bi211
                NBi211 -= decay_Bi211
                Nalpha_decay += decay_Bi211
            else:
                NPb207 += decay_Bi211
                NBi211 -= decay_Bi211
                NR_decay += decay_Bi211
    for v in range(NTl):
        if np.random.random() < p_Tl:
            decay_Tl+= 1
    NTl -= decay_Tl
    NPb207 += decay_Tl
    Nbeta_decay += decay_Tl


alpha_energy = Nalpha_decay * 5
R_energy = NR_decay * 7
beta_energy = Nbeta_decay * .5
Z_energy = NZ_decay * 10

print("The amount of Alpha decays were ", Nalpha_decay, "and the amount of Alpha energy is ", alpha_energy)
print("The amount of R decays were ", NR_decay, "and the amount of R energy is ", R_energy)
print("The amount of Beta decays were ", Nbeta_decay, "and the amount of Beta energy is ", beta_energy)
print("The amount of Z decays were ", NZ_decay, "and the amount of Z energy is ", Z_energy)
fig = plt.figure(figsize=(12,7))
ax = fig.add_subplot(111)
ax.plot(tpoints,Atpoints, c='b')
ax.plot(tpoints,Rnpoints, c='g') 
ax.plot(tpoints,Pb211points, c='r')
ax.plot(tpoints,Bi215points, c='c')
ax.plot(tpoints,Bi211points, c='k')
ax.plot(tpoints,Pb207points, c='y')
ax.plot(tpoints,Tlpoints, c='m')

ax.set_xlabel("Time (s)",size=20)   
ax.set_ylabel("Number of atoms",size=20)   

# increase the size of the numbers on the axes
ax.xaxis.set_tick_params(labelsize=20)

ax.yaxis.set_tick_params(labelsize=20)
plt.show()
```
